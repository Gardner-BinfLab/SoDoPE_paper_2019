```python

mABs = [Avastin, Humira,   Raptiva]
mABs_names = ['Avastin', 'Humira', 'Raptiva']
```

```python
items = ['Probability of solubility', 'Flexibility', 'GRAVY']
item = 0
testing = []
testing1 = []
testing2 = []
for i in mABs:
    testing.append(i['Full sequence'][item])
    testing1.append(i['V-set'][item])
    testing2.append(i['C1-set'][item])

df = pd.DataFrame({'mAB':mABs_names, 'V-set':testing1, \
              'Full sequence': testing, \
             'C1-set':testing2})
ax = df.sort_values(by="Full sequence").plot(kind="bar", figsize=(4, 4))

ax.set_xticklabels(df.sort_values(by="Full sequence")['mAB'])
plt.xticks(rotation=0,)
ax.set_xlabel('Commercial mAbs (light chain)')
ax.set_ylabel(items[item])
plt.savefig("prob_mAb.pdf", bbox_inches = 'tight', pad_inches = 0)
```
